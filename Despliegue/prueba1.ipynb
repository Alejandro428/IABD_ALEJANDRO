```python
6- Descarga de sprites:
Consulta el sprite (imagen) del Pokémon Pikachu y descárgalo utilizando curl.

Tip: Los sprites están en sprites.front_default

curl -X GET "https://pokeapi.co/api/v2/pokemon/pikachu" | jq -r '.sprites.front_default'
curl -X GET -o pikachu.png $(curl -s "https://pokeapi.co/api/v2/pokemon/pikachu" | jq -r '.sprites.front_default')

```
