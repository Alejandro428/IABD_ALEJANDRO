```python
# Paso 1
# De varios paises tenemos los siguientes datos del gatos en educación y su PIB.

#
#
#


import random
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from tabulate import tabulate


# Realiza lo siguiente:

# Crea una array de numpy de los datos.
# Obtén una "columna" llamada x_entrenamiento con los datos de entrada a la red neuronal. Ten en cuenta que debe ser una matriz ¿De cuantas columnas?
# Obtén una "columna" llamada y_entrenamiento con los datos de salida de la red neuronal. Ten en cuenta que debe ser una matriz ¿De cuantas columnas?

def get_datos():
    datos = [
        [  2.016,  56.180], [  4.797,  47.188], [  9.237,  57.689], [ 14.115,  43.703], [ 14.926,  59.102],
        [ 17.344,  65.960], [ 17.624,  45.743], [ 22.418,  13.575], [ 25.313,  68.437], [ 34.858, 147.153],
        [ 38.874,  25.396], [ 42.013,  74.390], [ 46.635,  98.933], [ 49.585, 116.070], [ 50.183, 138.555],
        [ 52.066, 139.366], [ 54.688, 150.096], [ 57.130, 156.143], [ 66.356, 119.758], [ 69.054, 139.081],
        [ 69.512, 128.722], [ 69.837, 152.651], [ 79.766, 148.231], [ 81.837, 137.863], [ 87.098, 217.289],
        [ 89.004, 168.649], [ 93.171, 163.105], [ 93.660, 200.476], [ 94.194, 150.440], [ 97.369, 173.205]
    ]
    return datos

datos = np.array(get_datos())

# 'datos' es una matriz de 2 columnas:
# columna 0 -> gasto en educación (entrada)
# columna 1 -> PIB (salida)

# Seleccionamos todas las filas y la primera columna de la matriz 'datos'
# Esto devuelve un array 1D (de una sola dimensión) con 30 elementos:
#    [2.016, 4.797, 9.237, 14.115, ...]
# Luego aplicamos .reshape(-1, 1) para convertirlo en una matriz 2D de 30 filas y 1 columna:
#    [[2.016],
#     [4.797],
#     [9.237],
#     [14.115],
#     ...]
# El -1 le dice a NumPy que calcule automáticamente cuántas filas hacen falta
# para que el array tenga la forma correcta (en este caso, 30 filas).
x_entrenamiento = datos[:, 0].reshape(-1, 1)  # columna 1 → entrada

# Seleccionamos todas las filas y la segunda columna (índice 1)
# De nuevo, reshape(-1, 1) convierte el vector 1D en una matriz columna (30x1)
# Esto es importante porque muchos modelos de aprendizaje automático esperan
# matrices 2D, no vectores 1D.
y_entrenamiento = datos[:, 1].reshape(-1, 1)  # columna 2 → salida

```

```python
# Paso 2
# Usando como base el siguiente código que crea y entrena un modelo de red neuronal.
#
def compile_fit(capas,x,y,epocas):
    np.random.seed(5)
    tf.random.set_seed(5)
    random.seed(5)

    model=Sequential()
    for index,neuronas_capa in enumerate(capas):
        if (index==0):
            model.add(Dense(neuronas_capa, activation='relu',input_dim=x.shape[1]))
        elif (index==len(capas)-1):
            model.add(Dense(neuronas_capa, activation='linear'))
        else:
            model.add(Dense(neuronas_capa, activation='relu'))
    model.compile(loss='mean_squared_error')

    history=model.fit(x, y,epochs=epocas,verbose=False)

    return model,history

# Modifica la función compile_fit para que el número de datos de entrada a la red sea el número de columnas de x, es decir modifica input_dim
# Modifica la función de activación de la última capa para que sea la función de activación linear.
# Añade un parámetros llamado epochs que sea el número de épocas a entrenar
# Crea una red neuronal y entrénala con:
# Capas: 5,10,15,10,1
# Épocas: 5000

capas = [5, 10, 15, 10, 1]

model,history = compile_fit(capas, x_entrenamiento, y_entrenamiento, 5000)
```

```python
# Calcula el error cuadrático para los siguientes datos sabiendo que el error es la resta entre el valor verdadero y el valor predicho
# , y luego elevándolo al cuadrado.

# Crea un array con los datos de entrada a predecir [ 2.01666708, 4.79734083, 9.23784581, 14.11529384 ] de forma que sea una columna
# (matriz con una cola columna) y llámalo x_paso3

# Datos de x
x_paso3 = np.array([2.01666708, 4.79734083, 9.23784581, 14.11529384]).reshape(-1, 1)

# Crea un array con los datos de salida verdaderos [ 56.18031474, 47.18848199, 57.68974048, 43.70348368] de forma que sea una columna
# (matriz con una cola columna) y llámalo y_true_paso3

# Datos de salida verdaderos
y_true_paso3 = np.array([56.18031474, 47.18848199, 57.68974048, 43.70348368]).reshape(-1, 1)

# Usando el modelo predice las salidas predichas y llámalo y_pred_paso3

y_pred_paso3 = model.predict(x_paso3, verbose=False)

# Resta el valor de y_true_paso3 a y_pred_paso3 y elévalo al cuadrado y llámalo error_paso3 :

error_paso3 = (y_true_paso3 - y_pred_paso3)**2

# Une los 4 arrays en una matriz donde cada columna sea cada uno de los arrays y llámala resultado_paso3

resultado_paso3 = np.column_stack([x_paso3, y_true_paso3, y_pred_paso3, error_paso3])

# Muestra el array con la función tabulate

print(tabulate(resultado_paso3, headers=['x', 'y_true', 'y_pred', 'Error^2']))
```

```python
# Paso 5
# Calcula una nueva media llamada Coeficiente de determinación o R². Lo bueno de esta nueva media es que da un valor entre -∞ y 1.
# Siendo 1 si no hay error, un 0 si es malo y negativo y es malísimo

# Para calcular el Coeficiente de determinación debemos hacerlo de la siguiente forma:

# Calcula la media de los valores verdaderos:

media_valores_verdaderos = np.mean(y_true_paso3)

# Calcula la suma de los errores (y_true-y_pred)²

suma_errores = np.sum(error_paso3)

# Calcula la suma de los errores entre la media y el valor verdadero (y_true-media_valores_verdaderos)²

suma_errores_media = np.sum((y_true_paso3 - media_valores_verdaderos)**2)

# Por último calcula el Coeficiente de determinación con la siguiente fórmula:

coeficiente = (1 - suma_errores/suma_errores_media)
```

```python
# Paso 6
# Usando el código de los pasos anteriores, crea una función llamada get_metricas_modelo(model,x,y_true)
# que retorne el Error cuadrático medio y el Coeficiente de determinación

def get_coeficiente_determinacion(y_true, y_pred):
    """
    Calcula el Coeficiente de Determinación (R^2)
    """
    # Media de los valores verdaderos
    media_valores_verdaderos = np.mean(y_true)

    # Suma de los cuadrados de los errores
    suma_errores = np.sum((y_true - y_pred) ** 2)

    # Suma total de cuadrados respecto a la media
    suma_valores_media = np.sum((y_true - media_valores_verdaderos) ** 2)

    # Coeficiente de determinación
    r2 = 1 - (suma_errores / suma_valores_media)
    return r2

def get_mse(y_true, y_pred):
    """
    Calcula el Error Cuadrático Medio (MSE) entre y_true y y_pred
    ESO CALCULA LA MEDIA DE LOS ERRORES ENTRE EL Y_TRUE Y EL Y_PRED
    """
    # Restamos los valores predichos de los valores reales, elevamos al cuadrado y hacemos la media
    mse = np.mean((y_true - y_pred) ** 2)
    return mse

def get_metricas_modelo(model, x, y_true):
    """
    Calcula el Error Cuadrático Medio (MSE) y el Coeficiente de Determinación (R^2)
    usando un modelo entrenado
    """
    y_pred = model.predict(x, verbose=False)
    mse = get_mse(y_true, y_pred)
    coeficiente_determinacion = get_coeficiente_determinacion(y_true, y_pred)
    return mse, coeficiente_determinacion



```
