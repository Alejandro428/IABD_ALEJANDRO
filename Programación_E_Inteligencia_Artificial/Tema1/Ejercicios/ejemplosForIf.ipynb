```python

def compilar_fit(array_capas, array_epocas, x, y):
    for i in range(len(array_capas)):
        capas_modelo = array_capas[i]   # Lista con neuronas por capa
        epocas = array_epocas[i]        # N√∫mero de √©pocas para este modelo

        model = Sequential()

        for index_capa, neuronas in enumerate(capas_modelo):
            if index_capa == 0:
                model.add(Dense(neuronas, activation='relu', input_dim=2))
            elif index_capa == len(capas_modelo) - 1:
                model.add(Dense(neuronas, activation='sigmoid'))
            else:
                model.add(Dense(neuronas, activation='relu'))

        model.compile(loss='mean_squared_error', optimizer='adam')
        print(f"\nüîÅ Entrenando modelo {i+1} con capas {capas_modelo} y {epocas} √©pocas...")
        model.fit(x, y, epochs=epocas, verbose=1)

# Llamada de ejemplo:
compilar_fit(
    array_capas=[
        [6, 12, 1],        # Modelo 1
        [8, 16, 1],        # Modelo 2
        [10, 20, 20, 1]    # Modelo 3
    ],
    array_epocas=[10, 20, 30],  # √âpocas correspondientes
    x=x,
    y=y
)
```
