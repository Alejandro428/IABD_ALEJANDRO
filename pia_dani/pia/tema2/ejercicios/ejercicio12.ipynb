```python
def compile_fit(capas, x, y, epochs=10):
    model = Sequential()
    
    for i, capa in enumerate(capas):
        if (i == 0):
            model.add(Dense(capa, activation='relu', input_dim=2))
        elif (i == len(capas) - 1):
            model.add(Dense(capa, activation='sigmoid'))
        else:
            model.add(Dense(capa, activation='relu'))

    model.compile(loss='mean_squared_error')

    inicio = time.time()
    model.fit(x, y, epochs=epochs, verbose=0)
    fin = time.time()
    
    tiempo = round(fin - inicio, 2)
    return model, tiempo
```

```python
resultados = []

for red, epocas in redes:
    model, tiempo = compile_fit(red, x, y, epochs=epocas)
    res1, res2 = model.predict(np.array([[4.9, 1.4], [6.3, 4.9]]))

    resultados.append([red, epocas, res1, res2, tiempo])
```
