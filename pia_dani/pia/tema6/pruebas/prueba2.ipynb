```python
from sklearn.datasets import load_iris
import numpy as np
import random
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelBinarizer
 
iris = load_iris()
x = iris.data              # shape (150,4)
y=iris.target

label_binarizer = LabelBinarizer()
label_binarizer.fit(range(max(y)+1))
y = label_binarizer.transform(y)

np.random.seed(5)
tf.random.set_seed(5)
random.seed(5)

model = Sequential()
model.add(Dense(6, activation='relu', input_dim=4))
model.add(Dense(12, activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(3, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(x, y, epochs=100)

# Predicciones (4 valores de entrada)
```

```python

pred = model.predict(np.array([[5.1, 3.5, 1.4, 0.2]]))
print("Probabilidades:", pred)
print("Clase predicha:", np.argmax(pred, axis=1))

pred = model.predict(np.array([[5.9, 3.0, 4.4, 1.3]]))
print("Probabilidades:", pred)
print("Clase predicha:", np.argmax(pred, axis=1))
```
