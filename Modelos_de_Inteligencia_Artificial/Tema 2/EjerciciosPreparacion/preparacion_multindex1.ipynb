```python
#1. Ventas Trimestrales: A partir de los siguientes datos, crea un DataFrame y 
#luego reestructúralo para que utilice un MultiIndex con 'Año' y 'Trimestre'. 
#Guarda el Dataframe en un fichero CSV, ya que lo necesitarás más adelante.

import pandas as pd
import numpy as np

datos_ventas = {
    'Año': [2022, 2022, 2023, 2023],
    'Trimestre': ['T1', 'T2', 'T1', 'T2'],
    'Ventas': [1500, 1800, 2100, 1900]
}

datos_trimestres_dataframe = pd.DataFrame(datos_ventas)

trimestre_multi = datos_trimestres_dataframe.set_index(['Año', 'Trimestre'])
print(trimestre_multi)
trimestre_multi.to_csv('informe_trimestre1.csv') # Guardamos el DataFrame como archivo .csv (texto separado por comas), sin índice
```

```python
# 2. Datos por Continente y País: A partir del siguiente diccionario, crea un DataFrame y establece un MultiIndex con 'Continente' y 'País'.

datos_poblacion = {
    'Continente': ['Europa', 'Europa', 'América', 'América'],
    'País': ['España', 'Francia', 'EEUU', 'Canadá'],
    'Poblacion': [47, 65, 331, 38]
}

datos_poblacion_dataframe = pd.DataFrame(datos_poblacion)
poblacion_multi = datos_poblacion_dataframe.set_index(['Continente', 'País'])
print(poblacion_multi)

```

```python
# 3. Selección por Nivel Externo: Recupera el DataFrame del ejercicio 1 ('Ventas Trimestrales')
#  desde el fichero CSV, selecciona y muestra todas las ventas correspondientes al año 2023 utilizando .loc[].

#lectura_datos_trimestre = pd.read_csv('informe_trimestre1.csv', index_col=['Año', 'Trimestre'])
lectura_datos_trimestre = pd.read_csv('informe_trimestre1.csv', index_col=['Año', 'Trimestre'])
# Usando index_col se convierten esas columnas seleccionadas en índices, por lo que facilita filtrar con ellas de manera sencilla
lectura_trimestre_filtrada = lectura_datos_trimestre.loc[2023]
# Como el primer índice es el año, usando loc, poniendo el año por el que queremos el filtro, funciona directamente
print(lectura_datos_trimestre)
```

```python
# 4. Selección por Nivel Interno: Partiendo del mismo DataFrame, selecciona y muestra las ventas de todos los 'T1' (primer trimestre) de todos los años.

# Ahora al querer filtrar por T1 (Primer trimestre) y este ser el segundo índice, hay que cambiar un par de cosas
# Ya no se puede usar simplemente el loc, ahora hay que usar el método xs, ya que queremos filtrar por un índice que no es el primero
# Este pide (parametro1 lo que queremos buscar, parametro2 el índice de los existentes sobre el que se quiere referenciar, y el parámetro3 por si se quiere mostrar también el dato con el que filtramos en el resultado 
# (en este caso, Trimestre))

resultado_trimestre = lectura_datos_trimestre.xs('T1', level='Trimestre', drop_level=False)
print(resultado_trimestre)
```

```python
# Desafío Práctico: Análisis de Sensores
# Escenario: Has recibido datos de sensores de temperatura. Tu objetivo es estructurarlos y realizar una consulta.

datos_sensores = {
    'Ubicacion': ['Norte', 'Norte', 'Sur', 'Sur', 'Norte', 'Sur'],
    'Sala': ['Cocina', 'Salón', 'Cocina', 'Salón', 'Cocina', 'Salón'],
    'Temperatura': [22, 21, 24, 23, 22.5, 23.5]
}

# Creación de Datos: Crea un DataFrame a partir del siguiente diccionario.

datos_sensores_dataframe = pd.DataFrame(datos_sensores)

# Estructuración: Convierte este DataFrame para que use un MultiIndex con 'Ubicacion' y 'Sala'.

sensores_multi = datos_sensores_dataframe.set_index(['Ubicacion', 'Sala'])

# Consulta: Usando el nuevo DataFrame estructurado, selecciona y muestra la temperatura de la 'Cocina' en la ubicación 'Sur'.

# El loc se puede usar para varios filtros al mismo tiempo, eso si, tienen que respetar el orden que tenga el índice del multi creado
# En este caso, el primero era ubicación y después sala, por lo que se puede hacer el filtro con ese orden
sensores_multi_filtrado = sensores_multi.loc[('Sur', 'Cocina')]
print(sensores_multi_filtrado)
```
