```python
import pandas as pd
import numpy as np

# 1. Downcasting Numérico: Crea un DataFrame con una columna 'ID_Usuario' con números del 
# 1 al 1000 (averigua cómo hacerlo usando "range"). Verifica su tipo de dato (probablemente int64) 
# y conviértelo al tipo entero más pequeño posible que pueda albergar esos valores (int16) usando
#  pd.to_numeric() con el argumento downcast.

datos_optimizacion = {
    'ID_Usuario': list(range(1, 1001))
}

datos_optimizacion_dataframe = pd.DataFrame(datos_optimizacion)

print(datos_optimizacion_dataframe['ID_Usuario'])
datos_optimizacion_dataframe.info()

datos_optimizacion_dataframe['ID_Usuario'] = pd.to_numeric(datos_optimizacion_dataframe['ID_Usuario'], downcast='integer')
datos_optimizacion_dataframe.info(memory_usage='deep')


```

```python
# 2. Optimización con Categorías: Crea un DataFrame con una columna 'Dia_Semana' que contenga 
# los días de la semana repetidos 500 veces. Compara el uso de memoria (usando .info(memory_usage='deep')) 
# antes y después de convertir la columna al tipo 'category'.

dias = ['Lunes', 'Martes', 'Miercoles', 'Jueves', 'Viernes', 'Sabado', 'Domingo']
dias_repetidos = dias * 500

dias_repetidos_dataframe = pd.DataFrame({"Dia_Semana": dias_repetidos})
dias_repetidos_dataframe.info(memory_usage='deep')

dias_repetidos_dataframe['Dia_Semana'] = dias_repetidos_dataframe['Dia_Semana'].astype('category')

dias_repetidos_dataframe.info(memory_usage='deep')



```
