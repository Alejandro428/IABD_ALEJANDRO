```python
import pandas as pd
import numpy as np

# 1. Filtrado de Emails: Crea un DataFrame con una columna de correos electrónicos. Utiliza el accesor 
# .str.contains() para filtrar y mostrar solo los emails que pertenecen al dominio '@gmail.com'.

datos_correos = {
    "Correo" : ["alejandro@gmail.com", "dani@gmail.com", "toni@hotmail.com"]
}

datos_correos_dataframe = pd.DataFrame(datos_correos)
print("Filas con '@gmail.com':\n", datos_correos_dataframe[datos_correos_dataframe['Correo'].str.contains('@gmail.com')])

```

```python
# 2. Desglose de Fechas: Partiendo de un DataFrame con una columna 'Fecha_Texto' 
# (con el formato 'DD/MM/YYYY'), usa .str.split() para separarla en tres nuevas 
# columnas: 'Día', 'Mes' y 'Año'.

df_fechas = {
    'Fecha_Texto': ['14/10/2025', '01/01/2024', '23/07/2023']
}

df_fechas_dataframe = pd.DataFrame(df_fechas)

df_fechas_dataframe[['Día', 'Mes', 'Año']] = df_fechas_dataframe['Fecha_Texto'].str.split('/', expand=True)

print(df_fechas_dataframe)
```

```python
# 3. Normalización de Nombres: Crea un DataFrame con una columna de nombres que contengan espacios en 
# blanco al inicio/final y mayúsculas/minúsculas inconsistentes (ej: [' Ana ', 'luis ', ' CARLOS']).
# Normaliza la columna convirtiendo todo a minúsculas y eliminando los espacios sobrantes con .str.lower() y .str.strip().

datos_nombres = {
    "Nombre": [' Ana ', 'luis ', ' CARLOS']
}

datos_nombres_dataframe = pd.DataFrame(datos_nombres)
datos_nombres_dataframe['Nombre'] = datos_nombres_dataframe['Nombre'].str.lower().str.strip()

print(datos_nombres_dataframe)
```

```python
#4. Limpieza de Números de Teléfono: Partiendo de un DataFrame con una columna 'Telefono' que contiene 
# números con guiones y espacios (ej: '600-11-22-33', '600 11 22 33'), usa .str.replace() para eliminar 
# tanto los guiones como los espacios, dejando solo los números.

datos_numeros = {
    "Numeros":['600-11-22-33', '600 11 22 33']
}

datos_numeros_dataframe = pd.DataFrame(datos_numeros)

datos_numeros_dataframe['Numeros'] = datos_numeros_dataframe['Numeros'].str.replace('-', '').str.replace(' ', '')
print(datos_numeros_dataframe)
```

```python
# 5. Desafío de Extracción de Datos: Tienes un DataFrame con una columna 'Info_Producto' 
# con un formato inconsistente como el del ejemplo. Tu objetivo es crear dos nuevas columnas limpias:
#'SKU': debe contener solo el número del producto (ej: '001', '002').
#'Producto': debe contener solo el nombre del producto, en minúsculas y sin espacios extra (ej: 'portátil', 'teclado').

datos_productos = {
    'Info_Producto': ['  SKU-001 : Portátil  ', 'sku-002: TECLADO', ' SKU-003: Ratón']
}

datos_productos_dataframe = pd.DataFrame(datos_productos)

datos_productos_dataframe[['SKU', 'Producto']] = datos_productos_dataframe['Info_Producto'].str.split(':', expand=True)
print(datos_productos_dataframe)

# El split en caso de utilizarlo sin  expand=True, genera una array con los dos lados separados
# Por lo que el str[1] se utiliza para quedarse con la parte que nos interesa en este caso
datos_productos_dataframe['SKU'] = datos_productos_dataframe['SKU'].str.split('-').str[1]
datos_productos_dataframe['Producto'] = datos_productos_dataframe['Producto'].str.lower().str.strip()
print(datos_productos_dataframe)
```
