```python
# Suponiendo que tienes un archivo llamado 'datos.csv'
# df = pd.read_csv('datos.csv')
# print(df.head()) # .head() muestra las primeras 5 filas

import pandas as pd

# 1 EXPORTAR A EXCEL
# Creamos un DataFrame simple con dos columnas y lo guardamos en un archivo de Excel
df = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df.to_excel('mi_primer_excel.xlsx', index=False) # Guardamos el DataFrame como archivo .xlsx (Excel), sin incluir el índice
```

```python

# 2 CARGAR Y GUARDAR EN CSV
# Creamos otro DataFrame con productos y precios, y lo exportamos como archivo CSV
df = pd.DataFrame({'Producto': ["Bananas", "Cholek"], 'Precio': [3, 4]})
df.to_csv('productos.csv', index=False) # Guardamos el DataFrame como archivo .csv (texto separado por comas), sin índice
```
