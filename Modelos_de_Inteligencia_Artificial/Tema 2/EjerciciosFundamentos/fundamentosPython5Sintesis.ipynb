```python
# Objetivo final: Crear CSV llamado informe_seleccion.csv
# El archivo deberá de incluir la siguiente información:
# - Debe incluir solo a los empleados que tengan más de 30 años
# - De estos, solo deben aparecer los que pertenecen a los departamentos de "Ventas" o "Marketing"
# - El archivo final solo debe contener las columnas "Nombre" y "Ciudad"
# - El archivo CSV no debe incluir el índice del DataFrame

import pandas as pd
import numpy as np

datos_empleados = {
    "Nombre": ["Ana", "Luis", "Marta", "Javier", "Sofía", "Carlos"],
    "Edad": [28, 34, 29, 42, 31, 45],
    "Ciudad": ["Valencia", "Madrid", "Barcelona", "Madrid", "Sevilla", "Valencia"],
    "Departamento": ["Marketing", "Ventas", "IT", "Ventas", "Marketing", "IT"]
}

df = pd.DataFrame(datos_empleados)
```

```python
# - Debe incluir solo a los empleados que tengan más de 30 años

# La condición la guardo para después reutilizarla más adelante
edad1condicion = df["Edad"] > 30
# El resultado es para mostrar directamente lo que sería equivalente ahora mismo
edad1resultado = df[df["Edad"] > 30]
print("Primer resultado:")
print(edad1resultado)
edad1resultado.to_csv('informe_seleccion1.csv', index=False) # Guardamos el DataFrame como archivo .csv (texto separado por comas), sin índice
```

```python
# - De estos, solo deben aparecer los que pertenecen a los departamentos de "Ventas" o "Marketing"

departamentos2condicion = (df["Departamento"] == "Ventas") | (df["Departamento"] == "Marketing")
departamentos2resultado = df[edad1condicion & departamentos2condicion]

print("Segundo resultado:")
print(departamentos2resultado)
departamentos2resultado.to_csv('informe_seleccion2.csv', index=False) # Guardamos el DataFrame como archivo .csv (texto separado por comas), sin índice
```

```python
# - El archivo final solo debe contener las columnas "Nombre" y "Ciudad"
# - El archivo CSV no debe incluir el índice del DataFrame

solocolumnas3resultado = departamentos2resultado[["Nombre", "Ciudad"]] #ESTO NO APLICA EN LA CABECERA LOS NOMBRES ADECUADAMENTE
print(solocolumnas3resultado)
solocolumnas3resultado.to_csv('informe_seleccion3.csv', index=False) # Guardamos el DataFrame como archivo .csv (texto separado por comas), sin índice
```

```python

# AMPLIACIÓN
# Hacer el ejercicio pero filtrando con los métodos .query() y .isin()

# Con Query
# Query sería igual que sql, pero poniendo los atributos con los nombres de las filas, y en la comparación 
# poner los valores de las columnas

resultado1 = df.query('Edad > 30 and (Departamento == "Ventas" or Departamento == "Marketing")')
resultado_final1 = resultado1[["Nombre", "Ciudad"]]
print(resultado_final1)
```

```python
# Con Isin
# Isin es parecido a la forma original, pero aquí no tienes que hacer comparaciones
# Directamente, puedes poner la fila que quieres y decir que columnas son las que quieres
# edad1condicion esta arriba, no hace falta redefinirlo, con isin no se puede atajar, es util para seleccion de columnas en una clave

ejemplo_isin_parte1 = df["Departamento"].isin(["Ventas", "Marketing"])
resultado_isinparte2 = df[edad1condicion & ejemplo_isin_parte1]
resultado_final2 = resultado_isinparte2[["Nombre", "Ciudad"]]
print(resultado_final2)
```
