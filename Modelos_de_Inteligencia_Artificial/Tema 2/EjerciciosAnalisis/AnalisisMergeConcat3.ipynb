```python
# 1. Uniendo Clientes y Pedidos: Crea dos DataFrames. El primero, df_clientes, debe tener las columnas 
# 'ID_Cliente' y 'Nombre'. El segundo, df_pedidos, debe tener 'ID_Cliente' y 'Producto'. Utiliza pd.merge() 
# para unirlos y obtener una tabla que muestre qué 'Nombre' de cliente pidió qué 'Producto'.
import pandas as pd

datos_clientes = {
    'ID_Cliente': [101, 102, 103],
    'Nombre': ['Ana', 'Luis', 'Marta']
}
df_clientes = pd.DataFrame(datos_clientes)

datos_pedidos = {
    'ID_Cliente': [101, 101, 102, 103],
    'Producto': ['Café', 'Tarta', 'Pastel', 'Té']
}
df_pedidos = pd.DataFrame(datos_pedidos)

union1 = pd.merge(df_clientes, df_pedidos, on='ID_Cliente', how='outer')
print(union1)
```

```python
# 2. Concatenando Datos Mensuales: Imagina que tienes dos DataFrames, ventas_enero y ventas_febrero, 
# ambos con las mismas columnas ('Producto', 'Cantidad'). Usa pd.concat() para unirlos en un único DataFrame 
# que contenga las ventas de ambos meses.

ventas_enero_datos = {
    'Producto': ['Café', 'Tarta', 'Té'],
    'Cantidad': [150, 40, 35]
}
ventas_enero = pd.DataFrame(ventas_enero_datos)

ventas_febrero_datos = {
    'Producto': ['Café', 'Tarta', 'Pastel'],
    'Cantidad': [130, 35, 25]
}
ventas_febrero = pd.DataFrame(ventas_febrero_datos)

concatenado2 = pd.concat([ventas_enero, ventas_febrero], ignore_index=True)
print(concatenado2)
```

```python
# 3. Reconstrucción de Datos de E-commerce:Tienes datos de un e-commerce distribuidos en tres tablas. 
# Tu misión es unificarlos para calcular el ingreso total por cada línea de pedido y luego identificar 
# el pedido que generó más ingresos.
# Crea los tres DataFrames a partir de los diccionarios de abajo.
# Usa pd.merge() para unir df_pedidos con df_detalles usando la clave 'Pedido_ID'.
# Realiza un segundo merge para unir el resultado anterior con df_productos usando la clave 'Producto_ID'.
# Crea una nueva columna llamada 'Ingreso_Total' que sea el resultado de multiplicar 'Cantidad' por 'Precio'.
# Agrupa el resultado por 'Pedido_ID' para sumar los ingresos de cada pedido y encontrar cuál fue el de mayor valor.

pedidos = {'Pedido_ID': [1, 2, 3], 'Cliente_ID': [101, 102, 101]}
detalles = {'Pedido_ID': [1, 1, 2, 3], 'Producto_ID': [10, 11, 10, 12], 'Cantidad': [2, 1, 5, 3]}
productos = {'Producto_ID': [10, 11, 12], 'Precio': [15.0, 50.0, 10.5]}

pedidos_df = pd.DataFrame(pedidos)
detalles_df = pd.DataFrame(detalles)
productos_df = pd.DataFrame(productos)

union_3_pedidos_detalles = pd.merge(pedidos_df, detalles_df, on='Pedido_ID', how='outer')
print(union_3_pedidos_detalles)

union_3_2_productos = pd.merge(union_3_pedidos_detalles, productos_df, on='Producto_ID', how='outer')
print(union_3_2_productos)

union_3_2_productos['Ingreso_Total'] = union_3_2_productos['Cantidad'] * union_3_2_productos['Precio']
print(union_3_2_productos)

agrupado_productos = union_3_2_productos.groupby('Pedido_ID')['Ingreso_Total'].agg(['sum', 'max'])
print(agrupado_productos)
```
