```python
# 1. Venta Media por Empresa: Utilizando el primer DataFrame de ejemplo 
# (el de las empresas GOOG, MSFT y FB), calcula la venta media (mean) 
# para cada compañía.

import pandas as pd
import numpy as np

data = {'Empresa': ['GOOG', 'GOOG', 'MSFT', 'MSFT', 'FB', 'FB'],
        'Persona': ['Sam', 'Charlie', 'Amy', 'Vanessa', 'Carl', 'Sarah'],
        'Ventas': [200, 120, 340, 124, 243, 350]}
df = pd.DataFrame(data)

venta_media = df.groupby('Empresa')['Ventas'].mean()
print(venta_media)
```

```python


# 2. Análisis Salarial por Departamento: Crea un nuevo DataFrame con dos columnas: 'Departamento' 
# (con valores como 'Ventas', 'IT', 'Marketing', repitiendo algunos) y 'Salario' (con valores numéricos).
#  Agrupa los datos por 'Departamento' y encuentra el salario máximo (max) en cada uno.

datos_salarios = {
    'Departamento': ['Ventas', 'IT', 'Marketing', 'IT', 'Ventas'],
    'Salario': [45000, 60000, 55000, 62000, 48000]
}
df_salarios = pd.DataFrame(datos_salarios)

analisis_salarial = df_salarios.groupby('Departamento')['Salario'].agg(['max'])
print(analisis_salarial)
```

```python
# 3. Agrupación por Múltiples Columnas: Crea un DataFrame que contenga 'Vendedor', 
# 'Producto' y 'Unidades_Vendidas'. Agrupa por ambas columnas, 'Vendedor' y 'Producto', 
# para calcular el total de unidades vendidas (sum) para cada combinación.

datos_multigrupo = {
    'Vendedor': ['Ana', 'Luis', 'Ana', 'Luis', 'Ana'],
    'Producto': ['A', 'A', 'B', 'A', 'B'],
    'Unidades_Vendidas': [10, 5, 8, 7, 12]
}
df_multigrupo = pd.DataFrame(datos_multigrupo)

total_unidades_vendidas = df_multigrupo.groupby(['Vendedor', 'Producto']).agg(['sum'])
print(total_unidades_vendidas)
```
