```python
# 1. Agregación Semanal: Crea una serie temporal con datos de ventas diarias para un periodo de 3 meses. 
# Después, usa .resample() para agregar los datos y obtener el total de ventas por semana ('W').

import pandas as pd
import numpy as np

# Generar fechas diarias para 90 días desde el 1 de enero de 2025
fechas = pd.date_range('2025-01-01', periods=90)

# Generar ventas aleatorias diarias entre 0 y 499
ventas_diarias = pd.Series(np.random.randint(0, 500, len(fechas)), index=fechas)

# Remuestrear a frecuencia semanal y sumar las ventas de cada semana
ventas_semanales = ventas_diarias.resample('W').sum()

# Mostrar los resultados
print("Ventas semanales:\n", ventas_semanales)
```

```python
# 2. Desafío - Crecimiento Mensual de Ventas:Una métrica clave para cualquier negocio es el crecimiento mes a mes (MoM). 
# Tu tarea es calcular este indicador.
# Partiendo de la serie temporal con datos de ventas diarias del ejercicio anterior, primero agrega los datos para obtener 
# el total de ventas de cada mes usando .resample().

ventas_mensuales = ventas_diarias.resample('ME').sum()
print("Ventas mensuales:\n", ventas_mensuales.head())


# Utiliza el método .shift() en los datos mensuales para crear una nueva serie que contenga las ventas del mes anterior.

# Ventas del mes anterior usando shift
ventas_mes_anterior = ventas_mensuales.shift(1)
print("Ventas del mes anterior:\n", ventas_mes_anterior.head())

# Calcula el crecimiento porcentual MoM usando la fórmula: ((ventas_actuales - ventas_anteriores) / ventas_anteriores) * 100.

crecimiento_mom = ((ventas_mensuales - ventas_mes_anterior) / ventas_mes_anterior) * 100
print("Crecimiento mensual (%):\n", crecimiento_mom.round(2).head())

# Esto te permitirá ver el porcentaje de crecimiento (o decrecimiento) de las ventas de un mes respecto al anterior.
```
