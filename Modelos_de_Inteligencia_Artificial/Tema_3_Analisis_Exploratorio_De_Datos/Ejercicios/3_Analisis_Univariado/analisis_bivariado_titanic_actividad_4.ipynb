```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('Titanic-Dataset.csv')

# Parte 1: Análisis con Python

# Usando el dataset del Titanic limpio de la actividad anterior (Actividad Práctica 1, Apartado 3), 
# vamos a investigar qué relaciones existen entre las variables y la supervivencia, comparando Python y Orange.

# Instrucciones:

# (Num vs Num): Crea un mapa de calor de correlación para las variables numéricas (`Age`, `Fare`, `SibSp`, `Parch`). 
# ¿Qué dos variables tienen la correlación más fuerte? ¿Es positiva o negativa?

subset = ['Age', 'Fare', 'SibSp', 'Parch']
corr_matrix = df[subset].corr(numeric_only=True)

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlación entre variables numéricas seleccionadas')
plt.show()
```

```python
# (Cat vs Cat): ¿Sobrevivieron más hombres o mujeres? Demuéstralo creando una tabla de contingencia (`pd.crosstab`) entre `Sex` y `Survived` y 
# luego visualízala con un `sns.countplot`.

tabla_contingencia = pd.crosstab(df['Sex'], df['Survived'])
print(tabla_contingencia)

# Sobrevivieron más mujeres (233) que hombres (109)
```

```python
# Visualizar la relación entre 'sex' y 'target'
sns.countplot(data=df, x='Sex', hue='Survived')
plt.legend(title='Sobrevive', labels=['No', 'Sí'])
plt.show()
```

```python
# (Num vs Cat): ¿La clase social (`Pclass`) influyó en la supervivencia? Compara la distribución de `Age`
#  para cada `Pclass` usando un gráfico de violín. ¿Ves alguna diferencia?

# Comparar distribución de colesterol entre sexos
sns.violinplot(data=df, x='Pclass', y='Age')
plt.show()

# Primera clase (Pclass = 1):
# Edades más variadas y presencia de personas mayores.
# Esto sugiere pasajeros con más recursos y familias acomodadas.

# Segunda clase (Pclass = 2):
# La mayoría son adultos jóvenes.
# Perfil más intermedio, típicamente familias de clase media.

# Tercera clase (Pclass = 3):
# Se observan muchos pasajeros jóvenes y niños.
# Era la clase de emigrantes y trabajadores con menos recursos.

# La supervivencia fue mayor en Pclass 1 y menor en Pclass 3.
# Sí, la clase social influyó en la supervivencia.
```

```python
# (Num vs Cat): ¿Pagó más la gente que sobrevivió? Compara la tarifa (`Fare`) frente a la supervivencia (`Survived`) 
# usando un diagrama de caja.

# Comparar distribución de colesterol entre sexos
sns.boxplot(data=df, x='Survived', y='Fare')
plt.show()

# Si, la gente que pagó más, sobrevivio
```
